## Handle Missing Values

```python
data.isnull().sum()
```

```python
# drop missing Values for Row
data=data.dropna()
```

```python
# drop missing values for columns
data.dropna(axis=1)
```

## Fill missing values:

```python
# fill the missing values (0)
data.fillna(0)
```

```python
# fill the nissing value for mean,max,min,std
data.fillna(data['Price'].mean())
data.fillna(data['Price'].max())
data.fillna(data['Price'].min())
data.fillna(data['Price'].std())
```

```python
# Use forward or backward fill:
data.fillna(method='ffill')
data.fillna(method='bfill')
```
